```python
import cv2
import os
from matplotlib import pyplot as plt

# Mendapatkan path lengkap ke file haarcascade_frontalface_default.xml
if "data" in dir(cv2):
    haarcascades_path = cv2.data.haarcascades
else:
    # Jika tidak ada cv2.data.haarcascades, cari path di direktori tempat skrip berjalan
    current_dir = os.path.dirname(os.path.abspath('__file__'))
    haarcascades_path = os.path.join(current_dir, 'haarcascades')

face_cascade_path = os.path.join(haarcascades_path, 'haarcascade_frontalface_default.xml')

# Membuat CascadeClassifier untuk deteksi wajah
face_cascade = cv2.CascadeClassifier(face_cascade_path)

# Fungsi untuk mendeteksi wajah dalam gambar dan menampilkannya di Jupyter Notebook
def detect_faces(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Deteksi wajah dalam gambar
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Menggambar kotak di sekitar wajah yang terdeteksi
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
    # Konversi gambar ke format yang sesuai untuk ditampilkan di Jupyter Notebook
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Menampilkan gambar dengan wajah yang terdeteksi menggunakan Matplotlib
    plt.figure(figsize=(8, 6))
    plt.imshow(img_rgb)
    plt.axis('off')  # Hilangkan sumbu x dan y
    plt.show()

# Path ke gambar yang ingin Anda deteksi
image_path = 'path_to_foto bahan tugas.jpg'  # Ubah path_to_image.jpg dengan path gambar Anda
detect_faces(image_path)
```
